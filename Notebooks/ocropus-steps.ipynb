Text Line Recognition=====================

There is currently only one text line recognizer included in OCRopus:- ocropus-lattices - compute recognition lattices by oversegmentation and character classificationThe `ocropus-lattices` command actually takes character recognizers and line segmenters andcombines them into a line recognizer.Additional recognizers are in preparation:- ocropus-hmm - HMM-based recognizer- ocropus-rnn - RNN-based recognizerThese produce similar files to `ocropus-lattices` but have their own internal recognizers.

```python
!ocropus-lattices 0020_0022/??????.png
```

The output from `ocropus-lattices` consists of a number of new files:- *.lattice - contains the recognition lattice (segmentation alternatives and character classes)- *.raw.txt - language-model free best interpretation of the input- *.rseg.png - color-coded segmentation of the input; colors are referenced by *.lattice

```python
!ls -1 0020_0022/010023.*
```

We can look at the segmentation output; to do so, we need to read the segmentation.In the segmentation, each RGB triple is interpreted as a 24 bit integer.The background is white.

```python
def read_segmentation(f):    seg = imread(f)    seg = array(255*seg,'i')    seg = 65536*seg[:,:,0]+256*seg[:,:,1]+seg[:,:,2]    seg *= (seg<32000)    return seg
```

To show characters in a distinctive way, we cycle between colors.

```python
rseg = read_segmentation("0020_0022/010023.rseg.png")figsize(32,12)subplot(121); imshow((rseg>0)*(rseg%3+1),cmap=cm.gist_stern)
```

The lattice file contains the recognition output.It is divided into segments and chars.Each `segment` line represents a range of labels in the raw segmentation.The numbers at the end represent the probability of a space being present or not present after the segment.The `chr` line represents the negative log probabilities and classes.

Language Modeling===================

The final output from the recognizer is computed via language modeling.Currently, the only language modeling code is in the separate `ocrofst` project,based on finite state transducers.Language models take the `*.lattice` files and output `*.txt` files and `*.cseg.png` files.

```python
!ocropus-lmodel 0020_0022/*.lattice
```

```python
!ls -1 0020_0022/010023.*
```

The `.cseg.png` file contains a segmentation that corresponds directly to the characters inthe output file.
