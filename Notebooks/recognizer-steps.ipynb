Loading the Image
=================

Select the input image here.

```python
fname = "010023.png"
#fname = "010052.png"
#fname = "010015.png"
```

Let's read the input image.

```python
image = ocrolib.read_image_gray(fname)
figsize(18,6); gray()
imshow(image)
```

Input images to the line segmenter should be black-on-white and only contain components from a single line.
The x-height should be somewhere between 15 and 30 pixels.  

The line recognizers will check whether line height and other aspects of the image are reasonable and otherwise
refuse to recognize the line (this can be overridden).

The line recognizer will also attempt to remove noise from the input.

The Segmenter
=================

Let's run the segmenter, just to see how well it is working by itself.

```python
reload(lineseg)
segmenter = lineseg.DPSegmentLine()
seg = segmenter.charseg(image)
morph.showlabels(seg)
```

The Character Recognizer
========================

Now we load the character recognition model.

```python
if "cmodel" not in dir():
    cmodel = ocrolib.load_component(ocrolib.ocropus_find_file("st1m.cmodel"))
```

We can actually run the character recognizer on a single character image.  Here is an example of that.

Note that inputs to the character recognizer are white-on-black.

```python
figsize(3,3)
cimage = (seg==1)[:,:50]
imshow(cimage)
cmodel.coutputs(cimage)
```

The Line Recognizer
====================

Here we construct the line recognizer from the segmenter and the character model.
There is also a space model.

```python
linerec = ocrolib.CmodelLineRecognizer(cmodel=cmodel,whitespace="space.model",
                                       maxrange=4,verbose=1,debug_cls=None,
                                       segmenter=segmenter)
linerec.use_ligatures = 0
linerec.maxcost = 10.0
linerec.nbest = 3
linerec.display = 0
linerec.norejects = 1
```

Now let's run the line recognizer.

```python
figsize(16,16)
linerec.recognize(image)
print linerec.bestpath()
imshow(linerec.image,interpolation='nearest')
```

You can see the segmentation graph and the recognized characters with the `ocropus-ngraphs --print` command.
The (vertical) numbers indicate the segment numbers in the raw segmentation.
The vertical position of each hypothesis indicates roughly how probable it is, with less probable
hypotheses being lower.

```python
grouper = linerec.getLattice()
grouper.saveLattice(open("test.lattice","w"))
!ocropus-ngraphs --print test.lattice
```

Let's look at the output from the line segmenter that the recognizer was actually using.

```python
figsize(16,16)
morph.showlabels(linerec.rseg)
```

And here we can see the individual characters and their associated outputs.

The special class "~" means a reject.

The values here are _costs_, i.e., negative log probabilities (the `coutputs` method for
the cmodel recognizer outputs posterior probabilities.)

```python
figsize(16,24)
for i,c in enumerate(linerec.chars[:256]):
    subplot(16,16,i+1)
    xticks([]); yticks([])
    xlabel("%s %.1f"%(c.outputs[0]))
    imshow(c.image,interpolation='nearest')
    
```
